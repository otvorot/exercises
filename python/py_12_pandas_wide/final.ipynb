```python
import pandas as pd
import re
data = pd.read_csv('data/ratings_movies.csv')
```

```python
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
data = data.drop('Unnamed: 0', axis=1)
```

```python
data['year release'] = data['title'].apply(get_year_release)
```

```python
null = data['year release'].info()
```

```python
min_rate = data[data['year release'] == 1999]
min_rate.groupby(by='title')['rating'].mean().sort_values(ascending=True)
```
